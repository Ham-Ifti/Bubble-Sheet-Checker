```python
import cv2
import numpy
from matplotlib import pyplot as plt
```

```python
image = cv2.imread("A4_images/3.jpeg", cv2.IMREAD_GRAYSCALE)
# Gaussian Blur
blur = cv2.GaussianBlur(image, (5, 5), 1)
# Canny Edge Detection
canny = cv2.Canny(blur, 30, 150)

# Dilate
kernel = numpy.ones((5, 5), numpy.uint8)
dilate = cv2.dilate(canny, kernel, iterations=1)

# Find Contours
contours, hierarchy = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Draw Contours
contour_image = cv2.drawContours(image.copy(), contours, -1, (0, 255, 0), 3)

# biggest contour
biggest_contour = max(contours, key=cv2.contourArea)

# Draw biggest contour
biggest_contour_image = cv2.drawContours(image.copy(), [biggest_contour], -1, (0, 255, 0), 5)

# Show images
plt.figure(figsize=(20, 20))
plt.subplot(1, 6, 1)
plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.subplot(1, 6, 2)
plt.imshow(blur, cmap="gray")
plt.title("Gaussian Blur")
plt.subplot(1, 6, 3)
plt.imshow(canny, cmap="gray")
plt.title("Canny Edge Detection")
plt.subplot(1, 6, 4)
plt.imshow(dilate, cmap="gray")
plt.title("Dilate")
plt.subplot(1, 6, 5)
plt.imshow(contour_image, cmap="gray")
plt.title("Contours")
plt.subplot(1, 6, 6)
plt.imshow(biggest_contour_image, cmap="gray")
plt.title("Biggest Contour")
plt.show()
```

```python
import os
images = []
for file in os.listdir("A4_images"):
    images.append(cv2.imread(os.path.join("A4_images", file), cv2.IMREAD_GRAYSCALE))

for image in images:
    blur = cv2.GaussianBlur(image, (5, 5), 0)
    canny = cv2.Canny(blur, 30, 100)
    kernel = numpy.ones((2, 2), numpy.uint8)
    dilate = cv2.dilate(canny, kernel, iterations=1)
    contours, hierarchy = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contour_image = cv2.drawContours(image.copy(), contours, -1, (0, 255, 0), 3)
    contours = sorted(contours, key=cv2.contourArea, reverse=True)
    biggest_contour = contours[0]
    biggest_contour_image = cv2.drawContours(image.copy(), [biggest_contour], -1, (0, 255, 0), 5)
    plt.imshow(biggest_contour_image, cmap="gray")
    plt.show()
```

```python
for image in images:
    blur = cv2.GaussianBlur(image, (5, 5), 0)    
    canny = cv2.Canny(blur, 30, 100)
    kernel = numpy.ones((2, 2), numpy.uint8)
    dilate = cv2.dilate(canny, kernel, iterations=1)
    contours, hierarchy = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contour_image = cv2.drawContours(image.copy(), contours, -1, (0, 255, 0), 3)
    contours = sorted(contours, key=cv2.contourArea, reverse=True)
    biggest_contour = contours[0]
    biggest_contour_image = cv2.drawContours(image.copy(), [biggest_contour], -1, (0, 255, 0), 5)
    mask = numpy.zeros(image.shape, dtype="uint8")
    cv2.drawContours(mask, [biggest_contour], -1, 255, -1)
    masked_image = cv2.bitwise_and(image, image, mask=mask)
    relt = cv2.minAreaRect(biggest_contour)
    box = cv2.boxPoints(relt).astype("int")
    angle = relt[-1]
    M = cv2.getRotationMatrix2D((masked_image.shape[1] // 2, masked_image.shape[0] // 2), angle, 1.0)
    rotated_image = cv2.warpAffine(masked_image, M, (masked_image.shape[1], masked_image.shape[0]))
    adaptive = cv2.adaptiveThreshold(rotated_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
    invert = cv2.bitwise_not(adaptive)
    contours, hierarchy = cv2.findContours(invert, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contour_image = cv2.drawContours(rotated_image.copy(), contours, -1, (0, 255, 0), 5)
    x, y, w, h = cv2.boundingRect(contours[0])
    cropped_image = rotated_image[y:y + h, x:x + w]
    adaptive = cv2.adaptiveThreshold(cropped_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
    dilate = cv2.dilate(adaptive, kernel, iterations=3)
    erode = cv2.erode(dilate, kernel, iterations=8)
    invert = cv2.bitwise_not(erode)
    plt.figure(figsize=(5,5))
    plt.subplot(1, 2, 1)
    plt.imshow(cropped_image, cmap="gray")
    plt.title("Cropped Image")
    plt.subplot(1, 2, 2)
    plt.imshow(invert, cmap="gray")
    plt.title("Final Image")
    plt.show()
        
    

```

```python
image = cv2.imread("A4_images/7.jpeg", cv2.IMREAD_GRAYSCALE)
# Gaussian Blur
blur = cv2.GaussianBlur(image, (5, 5), 0)

print(blur.shape)
plt.imshow(blur, cmap="gray")
plt.show()
```

```python
# Canny Edge Detection
canny = cv2.Canny(blur, 30, 100)

plt.imshow(canny, cmap="gray")
plt.show()
```

```python
# Dilate
kernel = numpy.ones((2, 2), numpy.uint8)
dilate = cv2.dilate(canny, kernel, iterations=1)

plt.imshow(dilate, cmap="gray")
plt.show()
```

```python
contours, hierarchy = cv2.findContours(dilate, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contour_image = cv2.drawContours(image.copy(), contours, -1, (0, 255, 0), 5)
print("Contours: ", len(contours))
plt.imshow(contour_image, cmap="gray")
plt.show()
```

```python
contours = sorted(contours, key=cv2.contourArea, reverse=True)
biggest_contour = contours[0]
biggest_contour_image = cv2.drawContours(image.copy(), [biggest_contour], -1, (0, 255, 0), 5)

plt.imshow(biggest_contour_image, cmap="gray")
plt.show()
```

```python
# mask the biggest contour
mask = numpy.zeros(image.shape, dtype="uint8")
cv2.drawContours(mask, [biggest_contour], -1, 255, -1)
masked_image = cv2.bitwise_and(image, image, mask=mask)

plt.imshow(masked_image, cmap="gray")
plt.show()
```

```python
relt = cv2.minAreaRect(biggest_contour)
box = cv2.boxPoints(relt).astype("int")
angle = relt[-1]
M = cv2.getRotationMatrix2D((masked_image.shape[1] // 2, masked_image.shape[0] // 2), angle, 1.0)
rotated_image = cv2.warpAffine(masked_image, M, (masked_image.shape[1], masked_image.shape[0]))

plt.imshow(rotated_image, cmap="gray")
plt.show()
```

```python
adaptive = cv2.adaptiveThreshold(rotated_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
invert = cv2.bitwise_not(adaptive)
contours, hierarchy = cv2.findContours(invert, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
contour_image = cv2.drawContours(rotated_image.copy(), contours, -1, (0, 255, 0), 5)
x, y, w, h = cv2.boundingRect(contours[0])
cropped_image = rotated_image[y:y + h, x:x + w]

plt.imshow(cropped_image, cmap="gray")
plt.show()

```
